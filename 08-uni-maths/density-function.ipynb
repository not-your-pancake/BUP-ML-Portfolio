```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = (1/3)x^2
def f(x):
    return (1/3) * x**2

# Generate x values
x_values = np.linspace(-3, 3, 500)

# Compute y values
y_values = f(x_values)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r"$f(x) = \frac{1}{3}x^2$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Graph of $f(x) = \\frac{1}{3}x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
