```python
import pandas as pd
```

```python
df = pd.read_csv('AQI Bangladesh.csv')

df.describe()
```

```python
# This returns a Series where names are the index and counts are the values
name_counts = df['city_name'].value_counts()
name_counts # 30 districts
```

```python
df_Dhaka = df[df['city_name'] == 'Dhaka']
df_Dhaka
```

```python
import pandas as pd

# 1. Convert to datetime if you haven't already
df_Dhaka['datetime'] = pd.to_datetime(df_Dhaka['datetime'])

# 2. Create the filter
mask = (df_Dhaka['datetime'] >= '2022-01-01') & (df_Dhaka['datetime'] <= '2025-12-31')

# 3. Apply it
df_filtered_Dhaka = df_Dhaka.loc[mask]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Ensure datetime is in the correct format
df_filtered_Dhaka['datetime'] = pd.to_datetime(df_filtered_Dhaka['datetime'])

# 2. Slice the last 1000 rows

# 3. Plot
plt.figure(figsize=(20, 6))
plt.plot(df_filtered_Dhaka['datetime'], df_filtered_Dhaka['pm10'], color='blue', linewidth=1.5)

# 4. Clean up the look
plt.title('Last 1000 pm10 Points')
plt.xlabel('Date & Time')
plt.ylabel('pm10')
plt.grid(True, alpha=0.1)
plt.gcf().autofmt_xdate() # Rotates dates so they don't overlap

plt.show()
```
